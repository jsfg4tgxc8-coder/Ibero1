```python
a = int(input("introduzca el primer numero"))
b = int (input("introduzca el segundo numero"))
op= input("escribe + ,-,* o /:")
if op=="+":
  print(a+b)
if op=="-":
  print(a-b)
if op=="*":
  print(a*b)
if op=="/":
  print(a/b)
```
